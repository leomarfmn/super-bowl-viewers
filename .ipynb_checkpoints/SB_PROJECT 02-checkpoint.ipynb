### **2. DATA CLEANING**

### 2.1. Dealing with null values

```python
df[df.isna().any(axis=1)]
```

```python
# It is easy to notice all the null values from the second file are from the year 2021. Those will be filled in
# manually next.

df[df['Date'].isna()]
```

```python
df.loc[df['Year'] == 2021,'Date'] = '2021-02-07'
df.loc[df['Year'] == 2021,'SB'] = 'LV'
df.loc[df['Year'] == 2021,'Attendance'] = '24835' #Covid restrictions. For more: https://www.scotsman.com/arts-and-culture/film-and-tv/how-many-fans-were-super-bowl-2021-why-nfl-supporters-were-allowed-stadium-amid-covid-restrictions-3127473
df.loc[df['Year'] == 2021,'Winner'] = 'Tampa Bay Buccaneers'
df.loc[df['Year'] == 2021,'QB  Winner'] = 'Tom Brady'
df.loc[df['Year'] == 2021,'Coach Winner'] = 'Bruce Arians'
df.loc[df['Year'] == 2021,'Winning Pts'] = '31'
df.loc[df['Year'] == 2021,'Loser'] = 'Kansas City Chiefs'
df.loc[df['Year'] == 2021,'QB Loser'] = 'Patrick Mahomes'
df.loc[df['Year'] == 2021,'Coach Loser'] = 'Andy Reid'
df.loc[df['Year'] == 2021,'Losing Pts'] = '9'
df.loc[df['Year'] == 2021,'MVP'] = 'Tom Brady'
df.loc[df['Year'] == 2021,'Stadium'] = 'Raymond James Stadium'
df.loc[df['Year'] == 2021,'City'] = 'Tampa'
df.loc[df['Year'] == 2021,'State'] = 'Florida'
df.loc[df['Year'] == 2021,'Point Difference'] = '22'
df.loc[df['Year'] == 2021,'Referee'] = 'Carl Cheffers'
df.loc[df['Year'] == 2021,'Geo Point 2D'] = '25.975833, -82.503056'
# TO DO
#df.loc[df['Year'] == 2021,'Geo Shape'] = 'LV'
```
