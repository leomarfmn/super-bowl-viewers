### **1. DATA COLLECTION AND DESCRIPTION**

```python
df1 = pd.read_csv('Super+Bowl+Commercials/superbowl_commercials.csv')
df1.sample(3)
```

```python
df1.info()
```

```python
df2 = pd.read_csv('super-bowl.csv', parse_dates = ['Date'], date_parser = pd.to_datetime, sep=';')
df2.sample(3)
```

```python
df2.info()
```

```python
df2['Year'] = df2['Date'].apply(lambda x: x.year)
df2.sample()
```

```python
df = pd.merge(df1, df2, on="Year", how = 'left')
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```
